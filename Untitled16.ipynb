```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from scipy.stats import zscore
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
print("--- Data Head ---")
print(df.head())
```

```python
print("\n--- Data Info ---")
print(df.info())
```

```python
print("\n--- Descriptive Statistics ---")
print(df.describe())
```

```python
if 'Unnamed: 0' in df.columns:
    df = df.drop('Unnamed: 0', axis=1)
    print("\n'Unnamed: 0' column dropped successfully.")
```

```python
print("\n--- Missing Values Count ---")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df, x_vars=['TV', 'radio', 'newspaper'], y_vars='sales', height=4, kind='reg')
plt.suptitle('Relationship between Ad Spend and Sales', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 4))
df.boxplot()
plt.title('Box Plot for Outlier Check')
plt.show()
```

```python
df_clean = df.copy()
```

```python
z_scores = zscore(df[['TV', 'radio', 'newspaper']])
abs_z_scores = np.abs(z_scores)
```

```python
filtered_entries = (abs_z_scores < 3).all(axis=1)
df_clean = df_clean[filtered_entries]
```

```python
print(f"\nOutliers Removed: {len(df) - len(df_clean)} rows.")
print(f"Original Shape: {df.shape}, Cleaned Shape: {df_clean.shape}")
```

```python
X = df_clean[['TV', 'radio', 'newspaper']]
y = df_clean['sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
print(f"\nTraining Data Points: {len(X_train)}")
print(f"Testing Data Points: {len(X_test)}")
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
ss_tot = np.sum((y_test - y_test.mean())**2)
ss_res = np.sum((y_test - y_pred)**2)
```

```python
r2 = 1 - (ss_res / ss_tot)
r2_percentage = r2 * 100
```

```python
print("\n--- Model Evaluation (Linear Regression) ---")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R2 Score): {r2:.4f}")
```
